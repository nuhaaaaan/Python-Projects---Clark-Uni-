```python
import pandas as pd

# Load the dataset
data_url = "https://data.giss.nasa.gov/gistemp/tabledata_v4/GLB.Ts+dSST.csv"
data = pd.read_csv(data_url, skiprows=1)  # Skipping the first row if it contains metadata

# Display the first few rows to understand the structure
print(data.head())
print(data.columns)
```

```python
# Filter for the years 1880 to 2021 and convert the "J-D" column to numeric
data = data[(data["Year"] >= 1880) & (data["Year"] <= 2021)]
data['J-D'] = pd.to_numeric(data['J-D'], errors='coerce')  # Convert to numeric, setting errors to NaN

# Drop any rows where "J-D" is NaN
data = data.dropna(subset=['J-D'])
```

```python
import matplotlib.pyplot as plt

# Plotting the annual global land and ocean temperature anomalies
plt.figure(figsize=(12, 6))
plt.plot(data["Year"], data["J-D"], label="Annual Anomaly", color="blue", linewidth=1.5)

# Step 4: Customize the Plot
plt.title("Global Land and Ocean Temperature Anomalies (1880-2021)")
plt.xlabel("Year")
plt.ylabel("Temperature Anomaly (Â°C)")
plt.legend()

# Display the plot
plt.show()
```

Analysis of Temperature Anomalies Over Time
The temperature anomaly data from 1880 to 2021 reveals distinct trends over three major periods:

1880 to 1950: During this period, temperature anomalies were relatively stable, fluctuating around the baseline with minor deviations. This stability suggests a relatively balanced global climate with no strong warming or cooling pattern.

1950s Onward: Around the 1950s, a clear upward shift in temperatures begins. This trend accelerates over time, especially as we move into the late 20th century. This rise in temperature anomalies aligns with the industrial boom and increased greenhouse gas emissions, marking the start of significant climate change signals.

Recent Decades (1980s to Present): The temperature anomalies show a pronounced increase in the past few decades, with almost all recent years showing values well above the baseline. This pattern strongly correlates with the global warming phenomenon, driven by human activity and industrial growth. The data underscores a persistent and accelerating rise in global temperatures, reflecting a critical phase in climate change that demands urgent attention.

Overall, this trend underscores the mounting impact of human activities on Earthâs climate, emphasizing the need for collective efforts to mitigate global warming.

```python
import pandas as pd

# Load the dataset
file_path = "Global food production data set 2019 - Seaborn analysis.csv"
data = pd.read_csv(file_path)

# Print the column names to confirm
print("Column names:", data.columns.tolist())

# Strip whitespace from column names
data.columns = data.columns.str.strip()

# Filter data for the year 2019
data_2019 = data[data["Year"] == 2019]

# Group by 'Area' and sum the 'Value' to get total food production
total_production = data_2019.groupby('Area')['Value'].sum().reset_index()

# Sort and select the top 10 countries by total food production
top10_countries = total_production.nlargest(10, 'Value')

# Display the result
print(top10_countries)
```

```python
# Set up the figure size and style
plt.figure(figsize=(12, 6))
sns.set_theme(style="whitegrid")

# Create a bar plot
sns.barplot(data=top10_countries, x='Area', y='Value', palette='viridis')

# Rotate x-axis labels for readability
plt.xticks(rotation=45)
plt.title('Top 10 Countries by Total Food Production in 2019')
plt.xlabel('Country')
plt.ylabel('Total Food Production (Value)')
plt.tight_layout()

# Show the plot
plt.show()
```

Economic Factors Influencing Food Production Rankings
When examining the factors that affect a country's standing in food production rankings, several economic elements come into play. Here are some key considerations:

Agricultural Infrastructure: Nations with robust agricultural infrastructureâsuch as advanced irrigation systems, modern machinery, and efficient transportation networksâtend to achieve higher levels of food production. This infrastructure facilitates the entire supply chain, from farming to market distribution.

Investment in Agriculture: Countries that channel significant investments into agricultural research and development often see improvements in productivity. Financial backing for farmers not only boosts yields but also encourages sustainable farming practices, leading to more resilient agricultural systems.

Economic Policies: Government policies and subsidies designed to support agriculture can greatly influence food production levels. Countries that create favorable trade agreements or provide financial incentives for specific crops can enhance production in those areas, making them more competitive globally.

Land Availability: The quantity of arable land is a critical factor in determining a nationâs capacity to produce food. Nations with extensive agricultural land can cultivate larger quantities of food compared to those with limited arable space.

Climate and Geography: A countryâs geographical and climatic conditions play a vital role in its agricultural success. Favorable weather patterns and suitable terrain allow for more efficient food production, giving those nations a distinct advantage.

Labor Availability: Access to a skilled labor force is essential in maximizing agricultural output. Countries with a substantial number of workers who possess expertise in agricultural practices are better equipped to optimize production levels.

Objective
This analysis of global food production distribution highlights not only the leading contributors but also the economic factors that shape agricultural output in various countries. Understanding these dynamics provides valuable insight into the complexities of food production on a global scale.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/Mobile cellular subscriptions data set.csv'
data = pd.read_csv(file_path, header=2)

# Check the column names to find the correct year column
print("Column names:", data.columns.tolist())

# Filter the relevant columns for 2019
# Select only the columns related to 'Country Name' and '2019'
data_2019 = data[['Country Name', '2019']].copy()

# Rename the columns for easier access
data_2019.columns = ['Country', 'Subscriptions_per_100']

# Display the first few rows to verify
print(data_2019.head())

# Drop any rows with missing values
data_2019.dropna(inplace=True)

# Create a histogram with a KDE overlay
plt.figure(figsize=(12, 6))
sns.histplot(data_2019['Subscriptions_per_100'], bins=30, kde=True, color='blue', stat="density")

# Add titles and labels
plt.title('Distribution of Mobile Cellular Subscriptions per 100 People in 2019', fontsize=16)
plt.xlabel('Mobile Subscriptions per 100 People', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Create a box plot
plt.figure(figsize=(12, 6))
sns.boxplot(x=data_2019['Subscriptions_per_100'], color='lightgreen')

# Add titles and labels
plt.title('Box Plot of Mobile Cellular Subscriptions per 100 People in 2019', fontsize=16)
plt.xlabel('Mobile Subscriptions per 100 People', fontsize=12)
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

Insights into Global Mobile Phone Penetration Trends
The analysis of mobile cellular subscriptions per 100 people in 2019 presents a compelling overview of global mobile phone penetration, along with significant regional disparities.

Overall Trends
Widespread Accessibility: The histogram indicates that many countries exhibit mobile subscription rates exceeding 100 per 100 people, suggesting that mobile phones have become an integral part of daily life globally. This trend illustrates that mobile technology is not merely a communication tool but a vital component of social and economic interaction.

Density Distribution: The Kernel Density Estimate (KDE) overlay demonstrates that a considerable number of nations cluster between 80 and 130 subscriptions per 100 people. This clustering signifies a trend toward saturation in mobile usage, particularly in developed countries, where access to mobile phones is almost universal.

Outliers Indicating Advanced Infrastructure: The box plot highlights several outliers, particularly in countries with exceptionally high subscription rates. These outliers often reflect nations with robust telecommunications infrastructure and market dynamics that favor mobile technology over traditional landline services.

Regional Disparities
Contrasts Between Developed and Developing Nations: There is a stark contrast between developed nations, such as those in North America and Europe, and developing countries, particularly in Africa and certain parts of Asia. This disparity can be attributed to differences in infrastructure development, economic resources, and technology accessibility.

Growth in Emerging Markets: In regions like Southeast Asia and parts of Latin America, a notable rise in mobile subscriptions is evident as economies develop and technology becomes more integrated into daily life. However, these areas still lag behind more developed regions, suggesting significant potential for growth in mobile penetration.

Urban-Rural Divide: Within individual countries, urban areas generally demonstrate higher mobile subscription rates compared to rural regions. This is largely due to better infrastructure, greater service availability, and more economic opportunities in urban settings, which facilitate higher mobile adoption rates.

Influence of Economic Factors: Economic stability plays a critical role in mobile phone penetration. Countries that invest in telecommunications infrastructure and provide incentives for mobile service providers tend to achieve higher subscription rates. Conversely, nations grappling with economic challenges may find it difficult to extend mobile access to their populations.

Conclusion
The findings on global mobile cellular subscriptions highlight the remarkable growth of mobile technology while also underscoring the persistent inequalities that exist across different regions and economic contexts. Recognizing these trends is essential for policymakers and businesses aiming to improve connectivity and drive economic growth through enhanced access to mobile technology. As mobile phones increasingly influence daily life, addressing these disparities will be crucial for ensuring that everyone benefits from the advantages of information and communication technologies on a global scale.
