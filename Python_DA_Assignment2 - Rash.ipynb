##Question 1

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = ('GLB.Ts+dSST.csv')
df = pd.read_csv(file_path, skiprows=1)
```

```python
df['J-D'] = pd.to_numeric(df['J-D'], errors='coerce')

years = df['Year']
temp_anomalies = df['J-D']
```

```python
plt.figure(figsize=(12, 6))
plt.plot(years, temp_anomalies, color='black')

plt.title('Global Temperature Anomalies (1880-2021)')
plt.xlabel('Year')
plt.ylabel('Temperature Anomaly')
plt.legend()
plt.grid(True)

plt.show()
```

##Task 4

**Early Years (1880-1920):** During this time, the temperature anomalies were pretty steady, mostly hanging between about -0.5C and -0.2C. There weren't many big changes; it was a stable period overall.

**Middle Years (1920-1970):** Starting around 1920, we see a gradual warming trend. The temperature anomalies shifted from being mostly negative to slightly positive. There was a peak around 1940 when the temperature hit about +0.2C, but then it stabilized a bit during the 1950s and 1960s.

**Recent Years (1970-2021):** After 1970, things really changed. Theres a clear upward trend where the temperature starts rising much faster. By 2000, the anomalies reached about +0.6C, and it just kept going up. By 2021, the temperature anomaly hit around +1.2C, which is the highest we've seen.

The warming has been speeding up, and the biggest increases have happened in the last 50 years. Weve gone from negative anomalies in the 1880s to positive ones over 1C in recent years, showing a big change in global temperatures. This trend is alarming, especially since it doesn't seem to be slowing down.

##Question 2

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = ('FAOSTAT_data_en_11-2-2024.csv')
food_data = pd.read_csv(file_path)
```

```python
production_2019 = food_data.loc[food_data['Year'] == 2019, ['Area', 'Value']]

country_totals = production_2019.groupby('Area', as_index=False)['Value'].sum()


top_producers = country_totals.sort_values(by='Value', ascending=False).head(10)
```

```python
plt.figure(figsize=(10, 8))
bar_plot = sns.barplot(x='Value', y='Area', data=top_producers, palette='viridis')
plt.title('Leading Countries in Food Production for 2019')
plt.xlabel('Total Food Production')
plt.ylabel('Country')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

##Task 4

Based on the graph showing the top 10 food-producing countries in 2019, we can see that economic factors really impact their rankings. Rich countries like the United States can spend more on better farming tools and technology, which helps them grow more food. They also have better roads and storage systems to handle their food supplies efficiently.

Wealthy nations invest in agricultural research, which leads to better crops and farming techniques. Countries with strong economies often give financial support to farmers, allowing them to manage their land better and keep food reserves. Developed countries also have easier access to international markets, meaning they can sell their food more easily abroad.

Population size matters too; countries like China and India have a big demand for food because of their large populations, and they also have a lot of workers available for farming. In richer countries, higher wages mean more use of machines in farming, while access to skilled workers improves agricultural practices. For example, China and India are putting money into their agriculture, the USA uses advanced technology, Brazil has a lot of land to farm, and countries like Indonesia and Russia are working on modernizing their agriculture as well.
